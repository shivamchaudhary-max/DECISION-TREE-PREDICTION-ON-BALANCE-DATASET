```python
import numpy as np

  # Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
balance=pd.read_csv('../input/balance/balance.csv')
```

```python
balance.isnull().sum()
```

```python
balance.head()
```

```python
(balance.set_axis(['Class','left_Weight','left_distance','right_weight','right_distance'], axis=1,inplace=True))
```

```python
balance.head()
```

```python
col=['left_Weight','left_distance','right_weight','right_distance']
x1=balance[col]
y1=balance.Class
```

```python
# Split dataset into training set and test set
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.3, random_state=1)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(x1_train,y1_train)
```

```python
#Predict the response for train dataset
y1_pred_train = clf.predict(x1_train)
```

```python
#Predict the response for test dataset
y1_pred = clf.predict(x1_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Train Accuracy:",metrics.accuracy_score(y1_train, y1_pred_train))
```

```python
# Model Accuracy, how often is the classifier correct?
print("Test Accuracy:",metrics.accuracy_score(y1_test, y1_pred))
```
